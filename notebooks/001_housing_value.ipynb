```python
import pandas as pd 
```

```python
fileIn = "/workspaces/house/data/house_value_history.xlsx"
```

```python
df = pd.read_excel(fileIn)
print(df.head())
```

```python
df['value'] = df['value'].str.replace('$', '').str.replace('K', '').astype(float) * 1000
print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(df['date'], df['value'])
plt.xlabel('Date')
plt.ylabel('House Value')
plt.title('House Value Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  
```

```python
fileOut = "/workspaces/house/data/house_value_history_cleaned.csv"
df.to_csv(fileOut, index=False)
```
