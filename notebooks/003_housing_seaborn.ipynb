### 002- Simple OLS model for housing data 

```python
# load libraties 
import pandas as pd 
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
fileIn = "../data/house_value_history_cleaned.csv"
```

```python
df = pd.read_csv(fileIn) 
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date_ordinal'] = df['date'].apply(lambda x: x.toordinal())

# Plot using the ordinal column
g = sns.lmplot(x='date_ordinal', y='value', data=df, aspect=1.5)

# 4. Format the x-axis back to dates
ax = g.axes[0, 0]
# Use a Formatter to turn ordinals back into strings
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Ensure labels don't overlap
plt.xticks(rotation=45)
plt.show()


```
