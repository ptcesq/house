### 002- Simple OLS model for housing data 

```python
# load libraties 
import pandas as pd 
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
fileIn = "../data/house_value_history_cleaned.csv"
```

```python
df = pd.read_csv(fileIn) 
df.head()
```

```python
# Convert string dates to datetime objects
df['date'] = pd.to_datetime(df['date'])
# Create a numerical time variable (e.g., months start date)
df['days'] = (df['date'] - df['date'].min()).dt.days
df.head()
```

```python
X = df['days']
y = df['value']
```

```python
# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)
```

```python
# Fit the OLS model
model = sm.OLS(y, X).fit()
# View the results
print(model.summary())
```
